# Training agents in the banana collector environment

This notebook shows how to train a deep Q-network to learn intelligent behavior in Unity's banana collector environment. 

```python
from unityagents import UnityEnvironment
from src.agent import Agent
from src.training import train_agent, show_scores_plot
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
mpl.rcParams['figure.dpi'] = 200
mpl.rcParams['figure.figsize'] = 10, 5
```

## Load environment
Loading the environment, saving the dimensions of the action and state space.

```python
env = UnityEnvironment(file_name="src/exec/Banana.app")
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
env_info = env.reset(train_mode=True)[brain_name]
action_size = brain.vector_action_space_size
state_size = brain.vector_observation_space_size
```

## Train agent
In this section, an example deep double Q-network is trained with an underlying network with two hidden layers with 512 units each. When the training is finished, the scores for each episode are returned and can be plotted.

This example only trains the network for 10 episodes. In order to achieve good results, consider training the network for around 2000 epochs. 

```python
agent = Agent(state_size, action_size, seed=0, double_ql=True, network_layers=[512,512])
scores = train_agent(env, brain_name, agent, n_episodes=10, epsilon_decay=0.997)
```

```python
show_scores_plot(scores)
```

It is also possible to save the plot (without displaying it) by specifying the `filename` parameter of the `show_scores_plot` function.
